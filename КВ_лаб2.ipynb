```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, Aer, execute
from qiskit.visualization import plot_histogram
circuit = QuantumCircuit(2, 2)
circuit.h(0)
circuit.cx(0, 1)
circuit.measure([0, 1], [0, 1])
circuit.draw()
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(circuit, backend, shots=1000)
result = job.result()
counts = result.get_counts(circuit)

plot_histogram(counts)
```

#1

```python
scheme1_1 = QuantumCircuit(2, 2)
scheme1_1.h(0)
scheme1_1.x(1)
scheme1_1.measure([0, 1], [0, 1])
scheme1_1.draw()
```

```python
# Выбираем симулятор состояний квантового компьютера
backend = Aer.get_backend('qasm_simulator')
# Запускаем квантовую схему на симуляторе
job = execute(scheme1_1, backend, shots=1000)
# Получаем результаты
result = job.result()
counts = result.get_counts(scheme1_1)

plot_histogram(counts)
```

```python
from qiskit.visualization import plot_bloch_multivector
cc = scheme1_1.copy()
cc.save_statevector()

vector = backend.run(cc).result().get_statevector()
plot_bloch_multivector(vector)
```
