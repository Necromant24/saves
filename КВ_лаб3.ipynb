```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, Aer, execute
from qiskit.visualization import plot_histogram

scheme1_4_1 = QuantumCircuit(2, 2)
scheme1_4_1.x(0)
scheme1_4_1.h(0)
scheme1_4_1.h(1)
scheme1_4_1.cx(0, 1)
scheme1_4_1.measure([1], [0])
scheme1_4_1.draw()
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(scheme1_4_1, backend, shots=1000)
result = job.result()
counts = result.get_counts(scheme1_4_1)

plot_histogram(counts)
```
