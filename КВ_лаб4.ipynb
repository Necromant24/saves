```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, Aer, execute
from qiskit.visualization import plot_histogram
import math

scheme2_1_a = QuantumCircuit(1, 1)
scheme2_1_a.rx(math.pi/1.25,0)
scheme2_1_a.measure([0], [0])
scheme2_1_a.draw()
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(scheme2_1_a, backend, shots=10000)
result = job.result()
counts = result.get_counts(scheme2_1_a)

plot_histogram(counts)
```
