```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, Aer, execute
from qiskit.visualization import plot_histogram
import math

scheme = QuantumCircuit(3, 3)
scheme.h(0)
scheme.h(1)
scheme.h(2)
scheme.barrier()
scheme.h(2)
scheme.cp(math.pi/4, 0, 2)
scheme.cp(math.pi/2, 1, 2)
scheme.h(1)
scheme.cp(math.pi/2, 0, 1)
scheme.h(0)
scheme.swap(0, 2)
scheme.barrier()
scheme.swap(0, 2)
scheme.h(0)
scheme.cp(-math.pi/2, 0, 1)
scheme.h(1)
scheme.cp(-math.pi/4, 0, 2)
scheme.cp(-math.pi/2, 1, 2)
scheme.h(2)
scheme.barrier()
scheme.measure(0, 0)
scheme.measure(1, 1)
scheme.measure(2, 2)

scheme.draw()
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(scheme, backend, shots=10000)
result = job.result()
counts = result.get_counts(scheme)

plot_histogram(counts)
```
